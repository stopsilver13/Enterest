```python
data = open('jamsil_main_seat_bts.csv', 'rt')
```

```python
data_split = []

for i in data :
    data_split.append(i.split(','))
    
print(data_split[:10])
```

```python
data_split[:10]
```

```python
for seat in data_split[10:]:
    if Block.objects.filter(name=seat[0]).exists():
        if seat[1] is '':
            level=SeatLevel.objects.get(name='')
        else:
            level=SeatLevel.objects.get(name=seat[1])
        Seat.objects.create(
            block=Block.objects.get(name=seat[0]),
            level=level,
            name=seat[2],
            row_real=seat[3],
            row=seat[4],
            col=seat[5],
        )
    else:
        block = Block.objects.create(
            section=Section.objects.get(pk=1),
            name=seat[0],
            coordinate='-',
        )
        if seat[1] is '':
            level=SeatLevel.objects.get(name='')
        else:
            level=SeatLevel.objects.get(name=seat[1])
        Seat.objects.create(
            block=block,
            level=level,
            name=seat[2],
            row_real=seat[3],
            row=seat[4],
            col=seat[5],
        )
    print(seat[0])
```
